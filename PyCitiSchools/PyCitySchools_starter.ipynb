```python
# Dependencies and Setup
import pandas as pd
import numpy as np 
from babel.numbers import format_currency, format_number, format_decimal, format_percent

# File to Load (Remember to Change These)
school_data_to_load = "Resources/schools_complete.csv"
student_data_to_load = "Resources/students_complete.csv"

# Read School and Student Data File and store into Pandas DataFrames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset.  
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
school_data_complete.head(20)
```

## District Summary

* Calculate the total number of schools

* Calculate the total number of students

* Calculate the total budget

* Calculate the average math score 

* Calculate the average reading score

* Calculate the percentage of students with a passing math score (70 or greater)

* Calculate the percentage of students with a passing reading score (70 or greater)

* Calculate the percentage of students who passed math **and** reading (% Overall Passing)

* Create a dataframe to hold the above results

* Optional: give the displayed data cleaner formatting

```python
total_schools = school_data["school_name"].count()
total_students = student_data["student_name"].count() #39170
total_budget = school_data["budget"].sum()
average_math = student_data["math_score"].mean()
average_reading = student_data["reading_score"].mean()

passing_math = student_data.loc[student_data["math_score"] > 70]
passing_reading = student_data.loc[student_data["reading_score"] > 70]

percentage_passing_math = passing_math["student_name"].count() * 100 / total_students #28356
percentage_passing_reading = passing_reading["student_name"].count() * 100 / total_students #32500

percentage_passing_both = pd.merge(passing_math, passing_reading)["student_name"].count() * 100 / total_students #23816
district_summary =   pd.DataFrame([
    {"Total Schools":total_schools, "Total Students": total_students, "Total Budget": total_budget, "Average Math Score": round(average_math,)
     , "Average Reading Score": round(average_reading,), "% Passing Math":f'{round(percentage_passing_math,2)}%'
     , "% Passing Reading":f'{round(percentage_passing_reading,2)}%', "% Overall Passing":f'{round(percentage_passing_both,2)}%'}
])
district_summary
```

## School Summary

* Create an overview table that summarizes key metrics about each school, including:
  * School Name
  * School Type
  * Total Students
  * Total School Budget
  * Per Student Budget
  * Average Math Score
  * Average Reading Score
  * % Passing Math
  * % Passing Reading
  * % Overall Passing (The percentage of students that passed math **and** reading.)
  
* Create a dataframe to hold the above results

```python
#Start the list of all schools and it's own School Type
school_summary = school_data_complete[["school_name","type"]]
school_summary = school_summary.rename(columns = {"type": "School Type"})
school_summary = school_summary.groupby("school_name").value_counts()
school_summary = pd.DataFrame(school_summary).reset_index()
school_summary = school_summary[["school_name","School Type"]]

#Count how many students are for each school
school_summary2 = school_data_complete[["school_name","student_name"]]
school_summary2 = school_summary2.rename(columns = {"student_name": "Total Students"})
school_summary2 = school_summary2.groupby("school_name").count().reset_index()

#Merge lists
school_summaryT = pd.merge(school_summary, school_summary2, on = "school_name")

#Get the total school budged for each school
school_summary3 = school_data_complete[["school_name","budget"]]
school_summary3 = school_summary3.rename(columns = {"budget": "Total School Budget"})
school_summary3 = school_summary3.groupby("school_name").value_counts()
school_summary3 = pd.DataFrame(school_summary3).reset_index()
school_summary3 = school_summary3[["school_name","Total School Budget"]]

#Merge lists
school_summaryT = pd.merge(school_summaryT, school_summary3, on = "school_name")

#Add a new column with the Total Student Budget assigned for each school
school_summaryT["Per Student Budget"] = school_summaryT["Total School Budget"] / school_summaryT["Total Students"]

#Get the average score for Math for each school
school_summary4 = school_data_complete[["school_name","math_score"]]
school_summary4 = school_summary4.rename(columns = {"math_score": "Average Math Score"})
school_summary4 = school_summary4.groupby("school_name").mean()
school_summary4 = round(pd.DataFrame(school_summary4).reset_index(),2)

#Merge lists
school_summaryT = pd.merge(school_summaryT, school_summary4, on = "school_name")

#Get the average score for Reading for each school
school_summary5 = school_data_complete[["school_name","reading_score"]]
school_summary5 = school_summary5.rename(columns = {"reading_score": "Average Reading Score"})
school_summary5 = school_summary5.groupby("school_name").mean()
school_summary5 = round(pd.DataFrame(school_summary5).reset_index(),2)

#Merge lists
school_summaryT = pd.merge(school_summaryT, school_summary5, on = "school_name")

#Get the the % Passing Math
school_summary6 = school_data_complete[["school_name","math_score"]]
school_summary6["math_score"] = school_summary6["math_score"].loc[school_summary6["math_score"] >= 70]
school_summary6 = school_summary6.dropna(how='any')
school_summary6 = school_summary6.groupby("school_name").count().reset_index() #3216
#Adding calculation for percenage to the main list
school_summaryT["% Passing Math"] = round(school_summary6["math_score"] * 100 / school_summaryT["Total Students"],2)

#Get the the % Passing Reading
school_summary7 = school_data_complete[["school_name","reading_score"]]
school_summary7["reading_score"] = school_summary7["reading_score"].loc[school_summary7["reading_score"] >= 70]
school_summary7 = school_summary7.dropna(how='any')
school_summary7 = school_summary7.groupby("school_name").count().reset_index()
#Adding calculation for percenage to the main list
school_summaryT["% Passing Reading"] = round(school_summary7["reading_score"] * 100 / school_summaryT["Total Students"],2)

#Adding calculation for Overall percenage to the main list
school_summary8 = school_data_complete[["school_name","math_score","reading_score"]]
school_summary8["math_score"] = school_summary8["math_score"].loc[school_summary8["math_score"] >= 70]
school_summary8["reading_score"] = school_summary8["reading_score"].loc[school_summary8["reading_score"] >= 70]
school_summary8 = school_summary8.dropna(how='any')
school_summary8 = school_summary8[["school_name","reading_score"]]
school_summary8 = school_summary8.rename(columns = {"reading_score": "All Passing"})
school_summary8 = school_summary8.groupby("school_name").count().reset_index()
#Adding calculation for percenage to the main list
school_summaryT["% Overall Passing"] = round(school_summary8["All Passing"] * 100 / school_summaryT["Total Students"],2)

#Adding currency and percentages to list
school_summaryT[["Total School Budget","Per Student Budget"]] = school_summaryT[["Total School Budget","Per Student Budget"]].apply(lambda series: series.apply(lambda x: format_currency(x, currency = "USD", locale="en_US")))
school_summaryT[["% Passing Math","% Passing Reading", "% Overall Passing"]] = school_summaryT[["% Passing Math","% Passing Reading", "% Overall Passing"]].apply(lambda series: series.apply(lambda x: f"{x}%"))

#Remove column name from first column
school_summaryT = school_summaryT.rename(columns = {"school_name": ""})

school_summaryT
```
